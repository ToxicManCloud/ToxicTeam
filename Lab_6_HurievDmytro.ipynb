```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = new_node
            self.rear = new_node
            return
        self.rear.next = new_node
        self.rear = new_node

    def dequeue(self):
        if self.front is None:
            return None
        temp = self.front
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return temp.data

    def print_n(self, n):
        current = self.front
        count = 0
        while current and count < n:
            print(current.data)
            current = current.next
            count += 1

# Приклад використання:
queue = Queue()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
queue.enqueue(4)
queue.enqueue(5)

print("Перші три елементи черги:")
queue.print_n(3)
```

Давайте оцінимо асимптотичну складність у середньому і в найгіршому випадку для операцій пошуку, вставки і видалення в черзі.

1.Пошук (search):
У черзі, реалізованій зв'язним списком, пошук виконується шляхом перегляду елементів від початку до кінця черги.
Складність пошуку в середньому і в найгіршому випадку становить O(n), де n - кількість елементів у черзі. Це тому, що в кожному випадку потрібно переглянути всі елементи черги.
2.Вставка (insert):
При вставці нового елемента до кінця черги, складність залишається O(1) у всіх випадках, оскільки немає необхідності переглядати всі елементи черги.
3.Видалення (delete):
При видаленні елемента з початку черги, складність також є O(1) у всіх випадках, оскільки потрібно просто оновити посилання на початок черги.
Проте, якщо потрібно видалити елемент з кінця черги у зв'язному списку, це займе O(n) часу, оскільки потрібно переглянути всі елементи для знаходження передостаннього елемента і оновлення посилання на останній елемент.
Отже, асимптотична складність операцій в черзі у середньому і в найгіршому випадку така:

Пошук: O(n)
Вставка: O(1)
Видалення з початку: O(1)
Видалення з кінця: O(n) (у зв'язному списку)

1.Стек - це структура даних, яка працює за принципом "останній прийшов, перший вийшов" (Last In, First Out, LIFO). Основні операції зі стеком включають додавання елементу на вершину стеку (push), видалення елементу з вершини стеку (pop) і перегляд верхнього елементу без видалення (peek).2.
Основна відмінність між стеком та чергою полягає в порядку, в якому елементи видаляються. У стеку елемент, який додався останнім, видаляється першим (LIFO), тоді як у черзі елемент, який додався першим, видаляється першим (FIFO).3.
Реалізація стека за допомогою масиву:
При реалізації стека за допомогою масиву, ми використовуємо одновимірний масив, де елементи можна додавати і видаляти з одного кінця - зверху. Індекс вершини стеку зазвичай використовується для визначення місця, де зберігається новий елемент. Основна перевага цього підходу - швидкість доступу до елементів за допомогою індексу. Однак ця реалізація обмежує розмір стеку масиво4.м.
Реалізація стека за допомогою зв’язаного списку:
У зв'язаному списку кожен елемент містить дані та посилання на наступний елемент у списку. У реалізації стека за допомогою зв'язаного списку новий елемент завжди додається до голови списку, а видалення елемента відбувається також з початку списку. Основна перевага цього підходу - гнучкість, так як розмір стеку не обмежений. Однак, в порівнянні з масивом, доступ до елементів може бути повільнішим через необхідність переходу по посилан5.нях.
Застосування стека та черги:
Стек використовується в алгоритмах обходу дерев, в алгоритмах зворотного польського запису, в управлінні викликами функцій та в управлінні пам'яттю.
Черга використовується в алгоритмах BFS (Breadth-First Search) для обходу графів, в системах керування процесами, в кешуванні даних та в імітації роботи різних пристроїв.
