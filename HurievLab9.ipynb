```python
def naive_string_search(text, pattern):
    n = len(text)
    m = len(pattern)
    for i in range(n - m + 1):
        j = 0
        while j < m and text[i + j] == pattern[j]:
            j += 1
        if j == m:
            print(f"–ü—ñ–¥—Ä—è–¥–æ–∫ –∑–Ω–∞–π–¥–µ–Ω–æ –Ω–∞ —ñ–Ω–¥–µ–∫—Å—ñ {i}")

# –¢–µ—Å—Ç—É–≤–∞–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–≤—ñ–ª—å–Ω–∏–º–∏ —Ä—è–¥–∫–∞–º–∏
text = "—Ü–µ –¥–æ–≤—ñ–ª—å–Ω–∏–π —Ä—è–¥–æ–∫ –¥–ª—è —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è"
pattern = "—Ä—è–¥–æ–∫"
naive_string_search(text, pattern)
```

–ù–∞—ó–≤–Ω–∏–π –∞–ª–≥–æ—Ä–∏—Ç–º –ø–æ—à—É–∫—É –ø—ñ–¥—Ä—è–¥–∫–∞ –º–∞—î –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å (n‚ãÖm), –¥–µ ùëõn ‚Äì —Ü–µ –¥–æ–≤–∂–∏–Ω–∞ —Ç–µ–∫—Å—Ç—É, –∞ ùëöm ‚Äì —Ü–µ –¥–æ–≤–∂–∏–Ω–∞ –ø—ñ–¥—Ä—è–¥–∫–∞. 

–£ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É, –¥–ª—è –∫–æ–∂–Ω–æ—ó –∑ n‚àím+1 –ø–æ–∑–∏—Ü—ñ–π —É —Ç–µ–∫—Å—Ç—ñ –º–∏ –ø–æ—Ä—ñ–≤–Ω—é—î–º–æ –¥–æ ùëöm —Å–∏–º–≤–æ–ª—ñ–≤, —Ç–æ–º—É –∑–∞–≥–∞–ª—å–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ—Ä—ñ–≤–Ω—è–Ω—å –±—É–¥–µ (n‚àím+1)‚ãÖm‚âàn‚ãÖm. 

```python
def z_function(s):
    n = len(s)
    Z = [0] * n
    l, r, k = 0, 0, 0
    for i in range(1, n):
        if i > r:
            l, r = i, i
            while r < n and s[r] == s[r - l]:
                r += 1
            Z[i] = r - l
            r -= 1
        else:
            k = i - l
            if Z[k] < r - i + 1:
                Z[i] = Z[k]
            else:
                l = i
                while r < n and s[r] == s[r - l]:
                    r += 1
                Z[i] = r - l
                r -= 1
    return Z

# –¢–µ—Å—Ç—É–≤–∞–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–≤—ñ–ª—å–Ω–∏–º —Ä—è–¥–∫–æ–º
s = "abcabcabc"
print(z_function(s))
```
