Обчислення факторілу O(n) 

Асимптотична складність цього алгоритму дорівнює O(n), оскільки кожна ітерація циклу for виконується один раз для кожного цілочисельного значення від 1 до n. Таким чином, кількість операцій зростає пропорційно з вхідним параметром n.
Таким чином, асимптотична складність рекурсивного алгоритму обчислення факторіалу - O(n).

Обчислення чисел Фібоначчі

Для оцінки асимптотичної складності цього алгоритму можна використовувати дерево рекурсії або метод розгортання рекурсії. За таким підходом, асимптотична складність рекурсивного алгоритму обчислення n-го числа Фібоначчі є вище експоненційної, а саме O(2^n), через те, що кожен виклик функції генерує два нових виклики, і такий процес повторюється до досягнення базових випадків.

Сортування злиттям

Алгоритм сортування злиттям має асимптотичну складність O(n log n), де n - кількість елементів у вихідному масиві. Цю складність можна оцінити за допомогою основної теореми рекурсії для алгоритмів сортування.

Основна теорема рекурсії для сортування злиттям стверджує, що складність алгоритму залежить від кількості операцій, необхідних для розбиття масиву на підмасиви та об'єднання відсортованих підмасивів.

Під час кожного кроку рекурсії масив розбивається навпіл, тобто поділяється на два підмасиви, поки не досягне базового випадку, коли розмір масиву стане одиницею. Цей процес злиття займає O(log n) кроків, де n - початковий розмір масиву.

Кожен злиття (об'єднання) підмасивів займає O(n) часу, оскільки кожен елемент масиву розглядається тільки один раз. Отже, загальна кількість операцій для злиття всіх підмасивів також є O(n).

Таким чином, асимптотична складність алгоритму сортування злиттям є добутком кількості операцій рекурсивного розбиття і кількості операцій злиття підмасивів, тобто O(n log n).
